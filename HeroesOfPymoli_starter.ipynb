## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)


#bins and labels
age_bins = [0, 10, 15, 20, 25, 30, 35, 40, 45]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
ages = pd.cut(purchase_data["Age"], age_bins, labels = age_labels)  #creating a cut
ages_df = pd.DataFrame(ages) #creating a new dataframe with the cut
ages_grouped = ages_df.groupby("Age")["Age"].count()  #grouping cut based off age and counting ages in each category
ages_grouped_df = pd.DataFrame(ages_grouped) #creating new data framed with the counted and grouped data
ages_grouped_df_name = ages_grouped_df.rename({"Age": "Total Count"}, axis = 1)   #renaming column


#percentages
age_percent = ages_grouped / purchase_data["Purchase ID"].count()  * 100
age_percent_df = pd.DataFrame(age_percent)
age_percent_df_name = age_percent_df.rename({"Age": "Percentage of Players"}, axis = 1)


#merging counts and percentages
age_stat_merge = pd.merge(ages_grouped_df_name, age_percent_df_name, on= "Age")

age_stat_merge


#numbers are off with my output- will have to figure out
```
